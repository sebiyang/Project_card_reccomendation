###카드명 & 카드설명 스크레이핑 

```python
#카드명 &카드설명 
import requests
from bs4 import BeautifulSoup

tag1 = []

for index in range(1,12):
     a = requests.get("https://card.kbcard.com/CRD/DVIEW/HCAMCXPRICAC0047?pageNo=1&cateIdx={}".format(index))
     soup = BeautifulSoup(a.text, 'html.parser')
     tags_1 = soup.select('a.linkDetail')
     
     for i in tags_1:
         name = i.select_one('h3.tit-dep4').get_text().strip() #를 텍스트만 빼와서 공백제거
         context = i.select_one('p').get_text().strip()
         tag1.append([name,context])
     
```

```python
import pandas as pd
import numpy as np

tag1 = pd.DataFrame(tag1)
```

```python
tag1.drop_duplicates(inplace=True)
tag1 = tag1.dropna(axis=0)
```

```python
tag2 = []

for index in numbers:
     a = requests.get("https://card.kbcard.com/CRD/DVIEW/HCAMCXPRICAC0076?mainCC=a&cooperationcode=0{}".format(index))
     soup = BeautifulSoup(a.text, 'html.parser')
     tags_2 = soup.select('div.cardTit')
     tags_3 = soup.select('ul.cardList1')
     
     for i in tags_2:
         name = i.select_one('h1.tit').get_text().strip() 
         tag2.append([name])
```

```python
tag3 = []

for index in numbers:
     a = requests.get("https://card.kbcard.com/CRD/DVIEW/HCAMCXPRICAC0076?mainCC=a&cooperationcode=0{}".format(index))
     soup = BeautifulSoup(a.text, 'html.parser')
     tags_3 = soup.select('ul.cardList1')
     
     for i in tags_3:
         discount =  i.select_one('p').get_text().strip()
         portion =  i.select_one('strong').get_text().strip()
         tag3.append([discount,portion])
```

```python
import pandas as pd
import numpy as np

tag2 = pd.DataFrame(tag2)
tag3 = pd.DataFrame(tag3)
tag2.columns = ['name']
tag3.columns = ['discount','portion']

tag4 = pd.concat([tag2,tag3],axis=1)
tag4
```

```python
tag_final= pd.merge(tag1,tag4, how ='right', on ="name")
```
